```python
# The function for getting the range of each csv column data in Numerical range and Categorical values  
def get_data_ranges(file_path):
    """
    Reads a CSV file and prints the range (min and max) of data for each column.
    
    Args:
        file_path (str): The path to the CSV file.
    """
    try:
        df = pd.read_csv(file_path)
        
        for column in df.columns:
            # For numerical data, calculate min and max
            if pd.api.types.is_numeric_dtype(df[column]):
                min_val = df[column].min()
                max_val = df[column].max()
                print(f"Column '{column}': Numerical Range [{min_val}, {max_val}]")
            
            # For categorical or string data, show unique values
            else:
                unique_values = df[column].unique()
                print(f"Column '{column}': Categorical/String Values {unique_values}")
                
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

# create a map for converting string to int

# Land cover ['Water Body' 'Forest' 'Agricultural' 'Desert' 'Urban']
# Land cover ['1' '2' '3' '4' '5']
# Soil Type ['Clay' 'Peat' 'Loam' 'Sandy' 'Silt']
# Soil Type ['1' '2' '3' '4' '5']

```python
import pandas as pd

# Define the mapping dictionaries for the columns
land_cover_map = {
    'Water Body': 1,
    'Forest': 2,
    'Agricultural': 3,
    'Desert': 4,
    'Urban': 5
}

soil_type_map = {
    'Clay': 1,
    'Peat': 2,
    'Loam': 3,
    'Sandy': 4,
    'Silt': 5
}

# Load your dataset
# Replace 'your_dataset.csv' with the name of your file
try:
    df = pd.read_csv('flood_risk_dataset_india.csv')
    
    # Check if the columns exist before mapping
    if 'Land Cover' in df.columns and 'Soil Type' in df.columns:
        # Apply the mappings to the specified columns
        df['Land Cover'] = df['Land Cover'].map(land_cover_map)
        df['Soil Type'] = df['Soil Type'].map(soil_type_map)
        
        # Save the new DataFrame to a new CSV file
        df.to_csv('mapped_dataset.csv', index=False)
        print("Columns successfully mapped and saved to 'mapped_dataset.csv'.")
    else:
        print("Error: The 'Land Cover' or 'Soil Type' columns were not found in the dataset.")
        
except FileNotFoundError:
    print("Error: 'your_dataset.csv' was not found. Please make sure the file is in the same directory as your script.")
except Exception as e:
    print(f"An error occurred: {e}")
```

# Normalizing the data

```python
# Adds a new function `normalize_and_save_params` to perform Min-Max scaling on numeric data in a CSV file.

import pandas as pd
import json

def normalize_and_save_params(input_csv_path, output_csv_path, params_file_path):
    """
    Normalizes a CSV, saves the normalized data, and saves the min/max params to a JSON file.
    """
    try:
        df = pd.read_csv(input_csv_path)
        normalization_params = {}

        for column in df.columns:
            if pd.api.types.is_numeric_dtype(df[column]):
                min_val = df[column].min()
                max_val = df[column].max()

                normalization_params[column] = {'min': float(min_val), 'max': float(max_val)}
                
                if max_val - min_val != 0:
                    df[column] = (df[column] - min_val) / (max_val - min_val)
                else:
                    df[column] = 0.0

        df.to_csv(output_csv_path, index=False)
        
        # Save the dictionary to a JSON file
        with open(params_file_path, 'w') as f:
            json.dump(normalization_params, f, indent=4)
            
        print(f"Data normalized and saved to '{output_csv_path}'")
        print(f"Normalization parameters saved to '{params_file_path}'")

    except FileNotFoundError:
        print(f"Error: One of the files was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage:
# normalize_and_save_params('data.csv', 'data_normalized.csv', 'normalization_params.json')
```
