```python
# 1. Load the data
df = pd.read_csv('flood_risk_dataset_india.csv')

# 2. Select only numeric columns (to avoid errors with text)
numeric_df = df.select_dtypes(include=['number'])

# 3. Create a summary table
range_info = pd.DataFrame({
    'Min': numeric_df.min(),
    'Max': numeric_df.max(),
    'Range (Diff)': numeric_df.max() - numeric_df.min()
})

print(range_info)
```
